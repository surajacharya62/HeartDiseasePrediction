```python
#For target assigning text categorical value
def target(value):
    if value == 1:
        return 'Having Heart Disease'
        
    else:
        return 'No Heart Disease'
    
#---------------------
#For gender assigning text categorical value
def gender(value):
    if value == 1:
        return 'Male'
    else:
        return 'Female'
    
#-----------------------
#For Bloodsugar assigning text categorical value
def BloodSugar(value):
    if value == 1:
        return "Having blood sugar"
    else:
        return "Not having blood sugar"
    
#-----------------------
#For RestingECG assigning text categorical value
def RestingECG(value):
    if value == 0:
        return 'Showing probable'
    elif value == 1:
        return 'Normal'
    else :
        return "Abnormality"

#-----------------------
#For ExcerciseInducedAngina assigning text categorical value
def ExcerciseInducedAngina(value):
    if value == 1:
        return 'Yes'
    else:
        return 'NO'
       
#-----------------------
#For Slope assigning text categorical value
def Slope(value):
    if value == 0:
        return 'Downward sloping'
    elif value == 1:
        return 'Remain flat'
    else :
        return "Upward sloping"

#-----------------------
#For Numberofbloodvessels assigning text categorical value
def Numberofbloodvessels(value):
    if value == 0:
        return '1 vessel'
    elif value == 1:
        return '2 vessels'
    else :
        return "3 vessels"

#-----------------------
#For Numberofbloodvessels assigning text categorical value
def Disorderinblood(value):
    if value == 1:
        return 'Normal'
    elif value == 2:
        return 'Defect in blood flow'
    else :
        return "reversible defect"
```

```python
#applying the function for assigning text categorical values to each individual columns
column_target = data_heart['target'].apply(target)
column_sex = data_heart['sex'].apply(gender)
column_fbs = data_heart['blood_sugar'].apply(BloodSugar)
column_restECG = data_heart['resting_ecg'].apply(RestingECG)
column_ExcerciseInducedAngina = data_heart['excercise_induced_angina'].apply(ExcerciseInducedAngina)
column_slope = data_heart['slope'].apply(Slope)
column_ca = data_heart['number_of_vessels'].apply(Numberofbloodvessels)
column_thal = data_heart['thal'].apply(Disorderinblood)
#sns.countplot(x=column_sex,hue=column_target)
#plt.title('Gender')
```

```python
# implementation the outlier handling using quartile percentile
numerical_data = ['resting_blood_pressure','cholestrol','max_heart_rate','depression_induced_by_excercise']  
def imputing_outlier(drop = False):
    for column in numerical_data:
        variable = data_heart[column]
        Quartile_1 = np.percentile(variable, 25.) 
        Quartile_3 = np.percentile(variable, 75.) 
        interquartile_range = Quartile_3-Quartile_1 
        outlier_step = interquartile_range * 1.5 
        outliers = variable[~((variable >= Quartile_1 - outlier_step) & (variable <= Quartile_3 + outlier_step))].index.tolist()  
        if not drop:
            print('No of outliers: '+ column + ' is', len(outliers))
        if drop:
            data_heart.drop(outliers, inplace = True, errors = 'ignore')
            print('No of outliers remove', len(outliers))

imputing_outlier()
```

To identify heart disease is one of the most critical tasks for doctors or any health analyst in the health sector. Heart is the most critical and complex organ of the human body. Early prediction of heart problems is main goal of this project. There are many studies have already been done and several machine learning algorithms are applied to predict heart disease. This project is also shares the same goal but with highest and most reliable prediction score with proper studied dataset taken from four different database from different health institutions. This project demonstrates that without using feature selection technique with KFold cross validation method can provide better accuracy for the cardiovascular related disease. Using Recursive Feature Elimination (RFE) method different models predicted comparatively lower score. Training and testing with five different models with cross validation method the two models- Decision Tree and Random Forest work well particularly with similar highest score of 97% while including all features. It is obvious that with is increasing in age there likely to have cardiovascular disease but in many cases early age of people are also getting similar kind of disease in modern days. Having higher data of elder citizen, the model can be biased for the young age of people. It may not be enough for the young age to generalize more. It is also necessary and may need more data for the young age people to make consistent with all ages of the group so that model can predict significantly.
There is limitation to get proper studied heart related data in which missing value, noisy data, extra feature can lie within the dataset which leads imbalance so, better prediction cannot be done as expected. In future, I aim to use deep learning algorithms that can be the more robust to such issues.
